# Đọc dữ liệu

```python
# đọc dữ liệu đã được xử lý và chuyển về các dataframe
import pandas as pd 

data_train=pd.read_csv(r'/content/Files/train_viet_news') 

data_test=pd.read_csv('/content/Files/test_viet_news')
```

# Đọc W2V

```python
import numpy as np
# Read embedding
word_dict = []
embeddings_index = {}
embedding_dim = 300
max_feature = len(embeddings_index) + 2

f = open("/content/Files/W2V_ner.vec",encoding='utf-8')
for line in f:
    values = line.split(' ')
    word = values[0] 
    word_dict.append(word)
    try:
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
    except Exception as e:
        pass
f.close()

print('Embedding data loaded')
```

# Tiền xử lý và tách câu

```python
# tách câu trong từng đoạn
# tập test
import nltk
import re
nltk.download('punkt')
index=[]
paras=[]
reg="[^\w\s]"
for i in data_test.index:
    data_test.original[i] = data_test.original[i].lower() # Chuyển về chữ thường
    data_test.original[i] = data_test.original[i].replace('\n', ' ') # Loại bỏ xuống đoạn
    data_test.original[i] = data_test.original[i].strip() # Loại bỏ khoảng trắng đầu và cuối đoạn
    paras_=nltk.sent_tokenize(data_test.original[i])
    for i in range(len(paras_)) :
       paras_[i] = re.sub(reg,'', paras_[i]) # Loại bỏ các ký tự đặc biệt
       paras_[i] = paras_[i].replace('   ',' ').replace('  ',' ')
       paras_[i]=paras_[i].strip()
    paras.append(paras_)
    index.append(i)
```

```python
# tách câu trong từng đoạn
# tập train
import nltk
nltk.download('punkt')
index_train=[]
paras_train=[]
reg="[^\w\s]"
for i in data_train.index:
    data_train.original[i] = data_train.original[i].lower() # Chuyển về chữ thường
    data_train.original[i] = data_train.original[i].replace('\n', ' ') # Loại bỏ xuống đoạn
    data_train.original[i] = data_train.original[i].strip() # Loại bỏ khoảng trắng đầu và cuối đoạn
    paras_ = nltk.sent_tokenize(data_train.original[i])
    for i in range(len(paras_)) :
       paras_[i] = re.sub(reg,'', paras_[i]) # Loại bỏ các ký tự đặc biệt
       paras_[i] = paras_[i].replace('   ',' ').replace('  ',' ')
       paras_[i]=paras_[i].strip()
    paras_train.append(paras_)
    index_train.append(i)
```

## Sentences Embedding

```python
import numpy as np 

# embedding (1 câu -> 1 vector)
paras_train_encode = []
for para in paras_train:
    sentence_encode=[]
    for sentence in para:
        words = sentence.split(" ")
        sentence_vec = np.zeros((300))
        for word in words:
            if word in embeddings_index.keys():
                sentence_vec+=embeddings_index[word] # Tính tổng các vector câu
            else:
                sentence_vec+=np.random.randn(300) # Nếu không có trong từ điển sẽ được khởi tạo ngẫu nhiên
        sentence_vec=sentence_vec/len(words) # Tính trung bình
        sentence_encode.append(sentence_vec)
    paras_train_encode.append(sentence_encode)

```
